This notebook illustrates the performance of VAE (VAE$_{clim \rightarrow clim + conv}$) as a function of latent space width in comparison to reference ANN and a linear version of it. In the second part the Table with VAEs test, validation and training MSEs and the AE, LR and reference ANN in the supporting material is produced

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

# varying latent space width and performance of VAE$_{clim\rightarrow clim +conv}$

```python
VAE_clim_clim_conv_BN2=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN2.csv')# VAE with latent space width of 2 nodes
VAE_clim_clim_conv_BN3=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN3.csv')# 3 nodes
VAE_clim_clim_conv_BN4=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN4.csv')# 4 nodes
# this is the VAE with 5 latent nodes used for the analysis in the paper
VAE_clim_clim_conv_BN5=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN5.csv')
###
VAE_clim_clim_conv_BN6=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN6.csv')# 6 nodes
VAE_clim_clim_conv_BN7=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN7.csv')# 7 nodes 
VAE_clim_clim_conv_BN8=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN8.csv')# 8 nodes

ref_ANN=pd.read_csv('perf_analy/reference_ANN.csv')#reference ANN based on Rasp et al. 2018

ref_ANN_lin=pd.read_csv('perf_analy/reference_ANN_linear_model.csv')#linear version of reference ANN

AE_clim_clim_conv=pd.read_csv('perf_analy/AED_clim_clim_conv.csv')#autoencoder decoder 
LR_clim_clim_conv=pd.read_csv('perf_analy/LR_clim_clim_conv.csv')# linear reference version of AE_clim_clim_conv
```

```python
VAE_clim_clim_conv_BN2.test_mse.values
```

```python
ref_ANN_lin
```

```python
VAE_test_mse=np.array([VAE_clim_clim_conv_BN2.test_mse.values,VAE_clim_clim_conv_BN3.test_mse.values,
                       VAE_clim_clim_conv_BN4.test_mse.values,VAE_clim_clim_conv_BN5.test_mse.values,
                       VAE_clim_clim_conv_BN6.test_mse.values,VAE_clim_clim_conv_BN7.test_mse.values,
                       VAE_clim_clim_conv_BN8.test_mse.values])#*np.ones((1,1))
```

```python
print(VAE_test_mse)
```

```python
plt.figure(1,(7,5))
plt.hlines(ref_ANN.test_mse,2,8,'blue',linestyles='--',label='Reference ANN')
plt.plot(np.arange(2,9),VAE_test_mse,'orange',marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$')
plt.hlines(ref_ANN_lin.test_mse,2,8,'black','--',label='Reference Linear model')
plt.legend(fontsize=15)
plt.grid(True)
plt.xlim([2,8])
plt.yticks(np.arange(0.14,0.24,0.02))
plt.ylabel('Test Data MSE',fontsize=16)
plt.xlabel('Latent Space Width',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('a)',fontsize=26,loc='left')
```

```python
VAE_val_mse=np.array([VAE_clim_clim_conv_BN2.val_mse.values,VAE_clim_clim_conv_BN3.val_mse.values,
                       VAE_clim_clim_conv_BN4.val_mse.values,VAE_clim_clim_conv_BN5.val_mse.values,
                       VAE_clim_clim_conv_BN6.val_mse.values,VAE_clim_clim_conv_BN7.val_mse.values,
                       VAE_clim_clim_conv_BN8.val_mse.values])
```

```python
VAE_val_mse
```

```python
plt.figure(1,(7,5))

plt.hlines(ref_ANN.val_mse,2,8,'blue',linestyles='--',label='Reference ANN')
plt.plot(np.arange(2,9),VAE_val_mse,'orange',marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$')
plt.hlines(ref_ANN_lin.val_mse,2,8,'black',linestyles='--',label='Reference linear model')
plt.grid(True)
#plt.legend(fontsize=12)
plt.xlim([2,8])
plt.yticks(np.arange(0.14,0.24,0.02))

plt.ylabel('Validation Data Mse',fontsize=16)
plt.xlabel('Latent Space Width',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('b)',fontsize=26,loc='left')

```

```python
VAE_train_mse=np.array([VAE_clim_clim_conv_BN2.train_mse.values,VAE_clim_clim_conv_BN3.train_mse.values,
                       VAE_clim_clim_conv_BN4.train_mse.values,VAE_clim_clim_conv_BN5.train_mse.values,
                       VAE_clim_clim_conv_BN6.train_mse.values,VAE_clim_clim_conv_BN7.train_mse.values,
                       VAE_clim_clim_conv_BN8.train_mse.values])
```

```python
plt.figure(1,(7,5))

plt.hlines(ref_ANN.train_mse,2,8,'blue',linestyles='--',label='reference ANN')
plt.plot(np.arange(2,9),VAE_train_mse,'orange',marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$')
plt.hlines(ref_ANN_lin.train_mse,2,8,'black',linestyles='--',label='reference linear model')
plt.grid(True)
plt.legend(fontsize=12)
plt.xlim([2,8])
plt.yticks(np.arange(0.14,0.24,0.01))

plt.ylabel('training data mse',fontsize=14)
plt.xlabel('latent space width',fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
plt.figure(1,(10,6))
plt.hlines(ref_ANN.test_mse,2,8,'blue',label='Reference ANN test')
plt.plot(np.arange(2,9),VAE_test_mse,color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE test')
plt.plot(np.arange(2,9),VAE_val_mse,'-.',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE val')
plt.plot(np.arange(2,9),VAE_train_mse,'--',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE train')

plt.hlines(ref_ANN_lin.test_mse,2,8,'black',label='Reference Linear Model test')
plt.vlines(5,0.13,0.23,color=([0,0.4,0.1]))
plt.legend(fontsize=15)
plt.grid(True)
plt.xlim([2,8])
plt.ylim(0.13,0.23)
plt.yticks(np.arange(0.13,0.24,0.01))
plt.ylabel(' MSE',fontsize=18)
plt.xlabel('Latent Space Width',fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```

```python
plt.figure(1,(10,6))
plt.hlines(ref_ANN.test_mse,2,8,'blue',label='Reference ANN test')
plt.plot(np.arange(2,9),VAE_test_mse,color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VED test')
plt.plot(np.arange(2,9),VAE_val_mse,'-.',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VED val')
plt.plot(np.arange(2,9),VAE_train_mse,'--',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VED train')

plt.hlines(ref_ANN_lin.test_mse,2,8,'black',label='Reference Linear Model test')
plt.vlines(5,0.13,0.23,color=([0,0.4,0.1]))
plt.legend(fontsize=15)
plt.grid(True)
plt.xlim([2,8])
plt.ylim(0.13,0.23)
plt.yticks(np.arange(0.13,0.24,0.01))
plt.ylabel(' MSE',fontsize=18)
plt.xlabel('Latent Space Width',fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```

```python
plt.figure(1,(10,6))
plt.hlines(ref_ANN.test_mse,2,8,'blue',label='Reference ANN test')
plt.plot(np.arange(2,9),VAE_test_mse,color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$ test')
plt.plot(np.arange(2,9),VAE_val_mse,'-.',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$ val')
plt.plot(np.arange(2,9),VAE_train_mse,'--',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$ train')

plt.hlines(ref_ANN_lin.test_mse,2,8,'black',label='Reference Linear Model test')
plt.vlines(5,0.13,0.23,color=([0,0.4,0.1]))
plt.legend(fontsize=15)
plt.grid(True)
plt.xlim([2,8])
plt.ylim(0.13,0.23)
plt.yticks(np.arange(0.13,0.24,0.01))
plt.ylabel(' MSE',fontsize=18)
plt.xlabel('Latent Space Width',fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```

```python
plt.figure(1,(10,6))
plt.hlines(ref_ANN.test_mse,2,8,'blue',label='Reference ANN test')
plt.plot(np.arange(2,9),VAE_test_mse,color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$ test')
plt.plot(np.arange(2,9),VAE_train_mse,'--',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$ train')

plt.hlines(ref_ANN_lin.test_mse,2,8,'black',label='Reference Linear Model test')
plt.vlines(5,0.13,0.23,color=([0,0.4,0.1]))
plt.legend(fontsize=15)
plt.grid(True)
plt.xlim([2,8])
plt.ylim(0.13,0.23)
plt.yticks(np.arange(0.13,0.24,0.01))
plt.ylabel(' MSE',fontsize=18)
plt.xlabel('Latent Space Width',fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```

# performance Tables on test, validation and  training data set

This section illustrates the data used to compile the MSE score table in the supporting material

```python
VAE_clim_clim_conv_perf=[r'VAE$_{clim \rightarrow clim + conv}$',
                                  np.round(VAE_clim_clim_conv_BN5.train_mse.values[0],3),
                                  np.round(VAE_clim_clim_conv_BN5.val_mse.values[0],3),
                                  np.round(VAE_clim_clim_conv_BN5.test_mse.values[0],3)]

AE_clim_clim_conv_perf=[r'AE$_{clim \rightarrow clim + conv}$',
                                  np.round(AE_clim_clim_conv.train_mse.values[0],3),
                                  np.round(AE_clim_clim_conv.val_mse.values[0],3),
                                  np.round(AE_clim_clim_conv.test_mse.values[0],3)]

LR_clim_clim_conv_perf=[r'LR$_{clim \rightarrow clim + conv}$',
                                  np.round(LR_clim_clim_conv.train_mse.values[0],3),
                                  np.round(LR_clim_clim_conv.val_mse.values[0],3),
                                  np.round(LR_clim_clim_conv.test_mse.values[0],3)]


reference_ANN_perf=['ANN',
                    np.round(ref_ANN.train_mse.values[0],3),
                    np.round(ref_ANN.val_mse.values[0],3),
                    np.round(ref_ANN.test_mse.values[0],3)]
```

```python
pd.DataFrame(np.array([VAE_clim_clim_conv_perf,AE_clim_clim_conv_perf,LR_clim_clim_conv_perf,reference_ANN_perf]
                      ,dtype='object'), columns=['Evaluated Network','Training MSE','Validation MSE','Test MSE'])
```

```python
VAE_clim_clim_conv_perf_paper=['VAE',
                                  np.round(VAE_clim_clim_conv_BN5.train_mse.values[0],3),
                                  np.round(VAE_clim_clim_conv_BN5.val_mse.values[0],3),
                                  np.round(VAE_clim_clim_conv_BN5.test_mse.values[0],3)]

AE_clim_clim_conv_perf_paper=['AE',
                                  np.round(AE_clim_clim_conv.train_mse.values[0],3),
                                  np.round(AE_clim_clim_conv.val_mse.values[0],3),
                                  np.round(AE_clim_clim_conv.test_mse.values[0],3)]

LR_clim_clim_conv_perf_paper=['LR',
                                  np.round(LR_clim_clim_conv.train_mse.values[0],3),
                                  np.round(LR_clim_clim_conv.val_mse.values[0],3),
                                  np.round(LR_clim_clim_conv.test_mse.values[0],3)]


reference_ANN_perf_paper=['ANN',
                    np.round(ref_ANN.train_mse.values[0],3),
                    np.round(ref_ANN.val_mse.values[0],3),
                    np.round(ref_ANN.test_mse.values[0],3)]
```
