```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import seaborn as sns
import pandas as pd
```

The following notebook was used to compute the density plots shown in Figure S11 in the supporting material

```python
VAE_clim_clim_conv_encoder=np.load('VAE_clim_clim_conv_encoder_pred_3_month_global.npy') #load the latent space data set
```

```python
# select a subset of latent space samples for the computation of KDEs
rand_data=np.random.choice(VAE_clim_clim_conv_encoder[:,1].size,50000,replace=False)
```

swap the order of the nodes:

original Node 1 $\rightarrow$ Node 3 in the paper

original Node 2 $\rightarrow$ Node 2 in the paper

original Node 3 $\rightarrow$ Node 4 in the paper

original Node 4 $\rightarrow$ Node 5 in the paper

original Node 5 $\rightarrow$ Node 1 in the paper

```python
VAE_clim_clim_conv_encoder_=VAE_clim_clim_conv_encoder[rand_data]

VAE_clim_clim_conv_encoder_swapped=np.nan*np.zeros(VAE_clim_clim_conv_encoder_.shape)
VAE_clim_clim_conv_encoder_swapped[:,0]=VAE_clim_clim_conv_encoder_[:,4]
VAE_clim_clim_conv_encoder_swapped[:,1]=VAE_clim_clim_conv_encoder_[:,1]
VAE_clim_clim_conv_encoder_swapped[:,2]=VAE_clim_clim_conv_encoder_[:,0]
VAE_clim_clim_conv_encoder_swapped[:,3]=VAE_clim_clim_conv_encoder_[:,2]
VAE_clim_clim_conv_encoder_swapped[:,4]=VAE_clim_clim_conv_encoder_[:,3]
```
