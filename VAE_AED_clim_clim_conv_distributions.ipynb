```python
## load some packages 
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import seaborn as sns
```

Load the saved latent spaces of VAE$_{clim \rightarrow clim + conv}$ and AE$_{clim \rightarrow clim + conv}$ for the SP test data. 

```python
VAE_clim_clim_conv_encoder=np.load('VAE_clim_clim_conv_encoder_pred_3_month_global.npy')
```

```python
AED_clim_clim_conv_encoder=np.load('AED_clim_clim_conv_encoder_3month.npy')
```

# Compress the two latent spaces to 2D with PCA

```python
VAE_pca=PCA(n_components=2).fit_transform(VAE_clim_clim_conv_encoder)
```

```python
AED_pca=PCA(n_components=2).fit_transform(AED_clim_clim_conv_encoder)
```

Calculate the explained variance of the first PCs of the latent space of VAE$_{clim \rightarrow clim + conv}$

```python
VAE_pca_ = PCA(n_components=2)
VAE_pca_.fit(VAE_clim_clim_conv_encoder)


VAE_pca_expl_var=np.sum(VAE_pca_.explained_variance_ratio_) 
```

```python
print(VAE_pca_expl_var)# print explained variance 
```

Calculate the explained variance of the first PCs of the latent space of AE$_{clim \rightarrow clim + conv}$

```python
AED_pca_ = PCA(n_components=2)
AED_pca_.fit(AED_clim_clim_conv_encoder)


AED_pca_expl_var=np.sum(AED_pca_.explained_variance_ratio_) 
```

```python
print(AED_pca_expl_var)# explained variance of first two leading PCs of latent space of AE_clim_clim_conv
```
