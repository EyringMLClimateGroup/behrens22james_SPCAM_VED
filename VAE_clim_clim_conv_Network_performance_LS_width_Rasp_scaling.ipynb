```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
#import xarray as xr
```

# varying latent space width and performance of VAE$_{clim\rightarrow clim +conv}$

```python
VAE_clim_clim_conv_BN2=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN2_PNAS_scaling.csv')# VAE with latent space width of 2 nodes
VAE_clim_clim_conv_BN3=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN3_PNAS_scaling.csv')# 3 nodes
VAE_clim_clim_conv_BN4=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN4_PNAS_scaling.csv')# 4 nodes
# this is the VAE with 5 latent nodes used for the analysis in the paper
VAE_clim_clim_conv_BN5=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN5_PNAS_scaling.csv')
###
VAE_clim_clim_conv_BN6=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN6_PNAS_scaling.csv')# 6 nodes
VAE_clim_clim_conv_BN7=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN7_PNAS_scaling.csv')# 7 nodes 
VAE_clim_clim_conv_BN8=pd.read_csv('perf_analy/VAE_clim_clim_conv_BN8_PNAS_scaling.csv')# 8 nodes

PNAS_ANN=pd.read_csv('perf_analy/reference_ANN_PNAS_scaling.csv')#reference ANN based on Rasp et al. 2018

PNAS_ANN_lin=pd.read_csv('perf_analy/reference_ANN_linear_model_PNAS_scaling.csv')#linear version of reference ANN

```

```python
VAE_clim_clim_conv_BN2.test_mse.values
```

```python
PNAS_ANN_lin
```

```python
VAE_test_mse=np.array([VAE_clim_clim_conv_BN2.test_mse.values,VAE_clim_clim_conv_BN3.test_mse.values,
                       VAE_clim_clim_conv_BN4.test_mse.values,VAE_clim_clim_conv_BN5.test_mse.values,
                       VAE_clim_clim_conv_BN6.test_mse.values,VAE_clim_clim_conv_BN7.test_mse.values,
                       VAE_clim_clim_conv_BN8.test_mse.values])#*np.ones((1,1))
```

```python
print(VAE_test_mse)
```

```python
plt.figure(1,(7,5))
plt.hlines(PNAS_ANN.test_mse,2,8,'blue',linestyles='--',label='Reference ANN')
plt.plot(np.arange(2,9),VAE_test_mse,'orange',marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$')
plt.hlines(PNAS_ANN_lin.test_mse,2,8,'black','--',label='Reference Linear model')
plt.legend(fontsize=15)
plt.grid(True)
plt.xlim([2,8])
plt.yticks(np.arange(0.004,0.008,0.0005))
plt.ylabel('Test Data MSE Rasp scaling',fontsize=16)
plt.xlabel('Latent Space Width',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('a)',fontsize=26,loc='left')
```

```python
VAE_val_mse=np.array([VAE_clim_clim_conv_BN2.val_mse.values,VAE_clim_clim_conv_BN3.val_mse.values,
                       VAE_clim_clim_conv_BN4.val_mse.values,VAE_clim_clim_conv_BN5.val_mse.values,
                       VAE_clim_clim_conv_BN6.val_mse.values,VAE_clim_clim_conv_BN7.val_mse.values,
                       VAE_clim_clim_conv_BN8.val_mse.values])
```

```python
VAE_val_mse
```

```python
plt.figure(1,(7,5))

plt.hlines(PNAS_ANN.val_mse,2,8,'blue',linestyles='--',label='Reference ANN')
plt.plot(np.arange(2,9),VAE_val_mse,'orange',marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$')
plt.hlines(PNAS_ANN_lin.val_mse,2,8,'black',linestyles='--',label='Reference linear model')
plt.grid(True)
#plt.legend(fontsize=12)
plt.xlim([2,8])
plt.yticks(np.arange(4E-3,8E-3,5E-4))

plt.ylabel('Validation Data Mse Rasp scaling',fontsize=16)
plt.xlabel('Latent Space Width',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('b)',fontsize=26,loc='left')

```

```python
VAE_train_mse=np.array([VAE_clim_clim_conv_BN2.train_mse.values,VAE_clim_clim_conv_BN3.train_mse.values,
                       VAE_clim_clim_conv_BN4.train_mse.values,VAE_clim_clim_conv_BN5.train_mse.values,
                       VAE_clim_clim_conv_BN6.train_mse.values,VAE_clim_clim_conv_BN7.train_mse.values,
                       VAE_clim_clim_conv_BN8.train_mse.values])
```

```python
plt.figure(1,(7,5))

plt.hlines(PNAS_ANN.train_mse,2,8,'blue',linestyles='--',label='reference ANN')
plt.plot(np.arange(2,9),VAE_train_mse,'orange',marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$')
plt.hlines(PNAS_ANN_lin.train_mse,2,8,'black',linestyles='--',label='reference linear model')
plt.grid(True)
plt.legend(fontsize=12)
plt.xlim([2,8])
plt.yticks(np.arange(4E-3,8E-3,5E-4))

plt.ylabel('training data mse Rasp scaling',fontsize=14)
plt.xlabel('latent space width',fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
plt.figure(1,(10,6))
plt.hlines(PNAS_ANN.test_mse,2,8,'blue',label='Reference ANN test')
plt.plot(np.arange(2,9),VAE_test_mse,color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE test')
plt.plot(np.arange(2,9),VAE_val_mse,'-.',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE val')
plt.plot(np.arange(2,9),VAE_train_mse,'--',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE train')

plt.hlines(PNAS_ANN_lin.test_mse,2,8,'k',label='Reference Linear Model test')
plt.vlines(5,4E-3,8E-3,color=([0,0.4,0.1]))
plt.legend(fontsize=15)
plt.grid(True)
plt.xlim([2,8])
plt.ylim(4.5E-3,7.5E-3)
plt.yticks(np.arange(4.5E-3,7.5E-3,5E-4))
plt.ylabel('MSE Rasp scaling',fontsize=18)
plt.xlabel('Latent Space Width',fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```

```python
plt.figure(1,(10,6))
plt.hlines(PNAS_ANN.test_mse,2,8,'blue',label='Reference ANN test')
plt.plot(np.arange(2,9),VAE_test_mse,color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$ test')
plt.plot(np.arange(2,9),VAE_val_mse,'-.',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$ val')
plt.plot(np.arange(2,9),VAE_train_mse,'--',color=([0,0.5,0.5]),marker='.',markersize=12,label=r'VAE$_{clim \rightarrow \ clim+conv}$ train')

plt.hlines(PNAS_ANN_lin.test_mse,2,8,'k',label='Reference Linear Model test')
plt.vlines(5,4E-3,8E-3,color=([0,0.4,0.1]))
plt.legend(fontsize=15)
plt.grid(True)
plt.xlim([2,8])
plt.ylim(4.5E-3,7.5E-3)
plt.yticks(np.arange(4.5E-3,7.5E-3,5E-4))
plt.ylabel('MSE Rasp scaling',fontsize=18)
plt.xlabel('Latent Space Width',fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```
